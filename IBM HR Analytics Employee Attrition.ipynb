**Data Cleaning**

```python
#Checking for missing values

df.isnull().sum().sort_values(ascending = False)
```

```python
#Checking for duplicate rows

df.duplicated().sum()
```

```python
#Checking for Columns with constant values

constant_columns = [col for col in df.columns if df[col].nunique() == 1]
print("Columns with constant values: ", constant_columns)
```

```python
#Dropping columns with constant values

df.drop(columns = constant_columns, inplace =  True)
```

```python
#Dropping columns with unnecessary informatin

drop_columns = ["EmployeeNumber", "MonthlyRate", "DailyRate"]
df.drop(columns = drop_columns, inplace = True)
df
```

**Exploratory Data Analysis and Data Visualization**

```python
#Check the number of employees who left the company and those who did not

df["Attrition"].value_counts()
```

```python
#Check the departments in the company

df["Department"].unique()
```

```python
#Check the number of employees in each department 

df["Department"].value_counts()
```

```python
#Check the number of employees in each department using bar plot

count = df["Department"].value_counts().reset_index()
count.columns = ["Department", "count"]
plt.figure(figsize = (5, 5))
sns.barplot(x = "Department", y = "count", data = count)
plt.title("Number of employees in each department")
plt.xlabel("Department")
plt.ylabel("Number of employees")
plt.show()
```

```python
#Checking the number of employees who left the company from each department

Left_employee = df[df["Attrition"] == "Yes"]
Attrition_Department = Left_employee["Department"].value_counts()
print("Number of employees who left the company from each department: ")
print(Attrition_Department)
```

```python
#Checking the number of employees who left the company from each department using piechart

plt.figure(figsize = (5, 5))
plt.pie(Attrition_Department, labels = Attrition_Department.index, autopct = "%1.1f%%", startangle = 90 )
plt.title("The number of employees who left the company from each department(as a percentage)")
plt.show()
```

```python
#Count the number of men and women 

df["Gender"].value_counts()
```

```python
#Checking the number of men and women who left 

left_employee = df[df["Attrition"] == "Yes"]
Attrition_gender = left_employee["Gender"].value_counts()
print("Number of employees who left the company from each gender: ")
print(Attrition_gender)
```

```python
#Checking the number of men and women who left using piechart

plt.figure(figsize = (5, 5))
plt.pie(Attrition_gender, labels = Attrition_gender.index, autopct = "%1.1f%%", startangle = 90 )
plt.title("The number of men and women who left(as a percentage)")
plt.show()
```

```python
#Checking the Distance From Home vs Attrition

plt.figure(figsize = (8, 6))
colors = {"Yes" : "blue", "No" : "red"}
plt.scatter(df.index, df["DistanceFromHome"], c = df["Attrition"].map(colors), alpha = 0.5)
plt.xlabel("Employee Index")
plt.ylabel("DistanceFromHome")
plt.title("Distance From Home vs Attrition")
plt.show()
```

```python
#Checking the Monthly Income vs Attrition

plt.figure(figsize = (8, 6))
colors = {"Yes" : "green", "No" : "red"}
plt.scatter(df.index, df["MonthlyIncome"], c = df["Attrition"].map(colors), alpha = 0.5)
plt.title("Monthly Income vs Attrition")
plt.xlabel("Employee Index")
plt.ylabel("Monthly Income")
plt.show()
```

```python
#The minimum age of an employees

df["Age"].min()
```

```python
#The maximum age of an employees

df["Age"].max()
```

```python
#Checking Employees vs Employees who left by age

left_age = df[df["Attrition"] == "Yes"]
count_age = df["Age"].value_counts().sort_index()
count_left = left_age ["Age"].value_counts().reindex(count_age .index, fill_value = 0)
plt.bar(count_age.index, count_age, color = "lightblue", label = "All employee")
plt.bar(count_left.index, count_left, color = "red", label = "Employees who left")
plt.title("Employees vs Employees who left by age")
plt.xlabel("Age")
plt.ylabel("Number of employees")
plt.legend()
plt.show()

```

```python
#Checking Years at company by attrition status

sns.boxplot(data = df, x = "Attrition", y = "YearsAtCompany", palette = "Set2")
plt.title("Years at company by attrition status")
plt.xlabel("Attrition")
plt.ylabel("Years at company")
plt.show()
```

```python
#Checking attrition by over time status

sns.countplot(data = df, x = "OverTime", hue = "Attrition")
plt.title("Attrition by over time status")
plt.xlabel("OverTime")
plt.ylabel("Attrition")
plt.show()
```
